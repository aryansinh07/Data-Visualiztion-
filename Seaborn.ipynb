```python
#Scatterplot 
#We can use relpot method bcoz it gives us both default as scatter but we can choose lineplot as well
```

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
#1.simple scatter plot
sns.relplot(df['total_bill'],df['tip'] , kind="scatter") #scatter is default so its not necessary for us to mention it....
```

```python
sns.relplot(x='total_bill' , y='tip' , data=df , kind='line')
```

```python
#Adding the hue semantics/feature
sns.relplot(df['total_bill'],df['tip'],hue=df['smoker'])
```

```python
#Adding the style feature/semantics
sns.relplot(df['total_bill'],df['tip'],hue=df['smoker'] , style=df['sex'])
```

```python
#Adding the size feature....
sns.relplot(df['total_bill'],df['tip'],hue=df['smoker'] , style=df['sex'] , size=df['size'])
```

```python
sns.relplot(df['total_bill'],df['tip'],hue=df['smoker'] , style=df['sex'] , size=df['size'] , sizes=(15,200))
```

# Heatmap

```python
df1 = sns.load_dataset('flights')
```

```python
df1
```

```python
x = df1.pivot_table(index='year' , columns='month' , values='passengers' , aggfunc='sum' )
```

```python
sns.heatmap(x)
```

# some imp features of Heatmap includee...
i) linewidth 
ii) annot
iii) fmt

```python
sns.heatmap(x , linewidth=0.5)
```

```python
sns.heatmap(x , linewidth=0.5 , annot=True , fmt='d')
```

# Clustermap

```python
sns.clustermap(x)
```
