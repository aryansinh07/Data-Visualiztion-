```python
import pandas as pd 
import numpy as np
```

```python
import matplotlib.pyplot as plt 
```

```python
age = [21,22,24,26,27,29,34,36]
```

```python
sd_salary=[35000,36000,44000,45000,60000,65000,75000,80000]
```

```python
Ds_salary=[40000,45000,50000,57000,59000,60000,79000,85000]
```

```python
plt.plot(age,sd_salary,label='Software Developer salary')
plt.plot(age,Ds_salary ,label='Data Scientist Salary')
plt.ylabel('Salary')
plt.xlabel('Engineers Age')
plt.title(' Salary vs Age')
plt.legend() #For getting the label 
plt.show()
```

```python
plt.plot(age,sd_salary,label='Software Developer salary' , color='r' , linestyle='-')
plt.plot(age,Ds_salary ,label='Data Scientist Salary' ,color='k' , linestyle='--')
plt.ylabel('Salary')
plt.xlabel('Engineers Age')
plt.title(' Salary vs Age')
plt.legend() #For getting the label 
plt.style.use("fivethirtyeigth")
plt.show()
#the parameters that we can add are color , linestyle , marker , label
```
